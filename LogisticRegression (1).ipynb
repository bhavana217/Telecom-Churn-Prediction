```
import numpy as np
import pandas as pd
import sklearn
sklearn.set_config(print_changed_only=True)
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import scale,StandardScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```
data = pd.read_csv("/content/Telecom_churn.csv")
```

```
data["Partner"].replace(["Yes","No"],[1,0],inplace=True)
data["Dependents"].replace(["Yes","No"],[1,0],inplace=True)
data["PhoneService"].replace(["Yes","No"],[1,0],inplace=True)
data["Churn"].replace(["Yes","No"],[1,0],inplace=True)
data["DeviceProtection"].replace(["Yes","No","No internet service"],[1,0,-1],inplace=True)
data["Contract"].replace(["Month-to-month","One year","Two year"],[0,1,2],inplace=True)
data["InternetService"].replace(["DSL","Fiber optic","No"],[2,1,0],inplace=True)
data["TechSupport"].replace(["Yes","No","No internet service"],[1,0,-1],inplace=True)
data["OnlineBackup"].replace(["No","Yes","No internet service"],[0,1,-1],inplace=True)
data["StreamingMovies"].replace(["No","Yes","No internet service"],[0,1,-1],inplace=True)
data["PaperlessBilling"].replace(["Yes","No"],[1,0],inplace=True)
data["gender"].replace(["Female","Male"],[0,1],inplace=True)
data["StreamingTV"].replace(["No","Yes","No internet service"],[0,1,-1],inplace=True)
data["MultipleLines"].replace(["Yes","No","No phone service"],[1,0,-1],inplace=True)
data["OnlineSecurity"].replace(["No","Yes","No internet service"],[0,1,-1],inplace=True)


```

```
data.corr()
```

```
data["TotalCharges"].replace([' '],[0],inplace=True)
data["TotalCharges"] = data["TotalCharges"].astype("float64")
data1 = data[data['TotalCharges'] != 0] 
m1 = data1['TotalCharges'].mean()
data['TotalCharges'].replace([0],[m1],inplace = True)
```

```
x = data.drop(['customerID','PhoneService','OnlineBackup','TechSupport','DeviceProtection','StreamingMovies','PaymentMethod','Dependents','Partner','gender','Churn'],axis = 1)
y = data.Churn
```

```
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)
```

```
from sklearn.linear_model import LogisticRegression
```

```
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
```

```
lr.score(x_test,y_test)
```

```
y_pred = lr.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
confusion_matrix(y_test,y_pred)
```

```
print(classification_report(y_test,y_pred))
```
